```python
#Importing necessary Packages for model building

# For logistic regression
import statsmodels.api as sm
from sklearn.linear_model import LogisticRegression

# For Recursive Feature elimination (RFE)
from sklearn.feature_selection import RFE

# Metrics import for model evaluation
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import roc_curve

# For defining the colors for the Matshow
from matplotlib.colors import ListedColormap

# Check for the VIF values of the feature variables 
from statsmodels.stats.outliers_influence import variance_inflation_factor

# For AUC-ROC curve
from sklearn import metrics
```

```python
logm1 = sm.GLM(y_train,(sm.add_constant(X_train)), family = sm.families.Binomial())

logm1.fit().summary()
```

#### Some features have high p value. Using Recursive Feature Elimination Technique to select appropriate features for the model

```python
logreg = LogisticRegression(random_state = 100)
```

```python
rfe = RFE(logreg, 30)             
rfe = rfe.fit(X_train, y_train)
```

```python
rfe.support_
```

```python
#extracting the feature names, rfe support and rank into a list
list(zip(X_train.columns, rfe.support_, rfe.ranking_))
```

```python
#extracting column names that support rfe
col = X_train.columns[rfe.support_]
```

```python
#building the model 
X_train_rfe = sm.add_constant(X_train[col])
logm2 = sm.GLM(y_train, X_train_rfe , family = sm.families.Binomial())
res = logm2.fit()
res.summary()
```

```python
#calculating variable inflation factor
vif = pd.DataFrame()
vif['Features'] = col
vif['VIF'] = [variance_inflation_factor(X_train[col].values, i) for i in range(X_train[col].shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

#### The VIF is less than 2 hence theres no need to drop any features

```python
#making predictions on train data with random treshold as 0.5
y_train_pred = res.predict(X_train_rfe)

y_train_pred = pd.DataFrame({'Y': y_train, 'Y_Pred': y_train_pred})
y_train_pred['thresh_0.5'] = y_train_pred['Y_Pred'].map(lambda x: 1 if x > 0.5 else 0)

y_train_pred.head()
```

```python
#defining a function to plot ROC curve
def draw_roc(actual, probs):
    fpr, tpr, thresholds = metrics.roc_curve(actual, probs, drop_intermediate = False)
    
    auc_score = metrics.roc_auc_score( actual, probs )
    
    plt.figure(figsize=(5, 5))
    
    plt.plot( fpr, tpr, label='ROC curve (area = %0.2f)' % auc_score )
    plt.plot([0, 1], [0, 1], 'k--')
    
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic example')
    plt.legend(loc="lower right")
    plt.show()

    return None
```

```python
#calculting true positive and false positive rates
fpr, tpr, thresholds = metrics.roc_curve(y_train_pred.Y, y_train_pred.Y_Pred, 
                                         drop_intermediate = False)
```

```python
#plotting ROC curve
draw_roc(y_train_pred.Y, y_train_pred.Y_Pred)
```

#### Area under ROC is good

```python
#making predictions for various treshold values
numbers = np.arange(0, 1, 0.1)

for i in numbers:
    y_train_pred['thresh_' + str(i.round(2))]= y_train_pred['Y_Pred'].map(lambda x: 1 if x > i.round(2) else 0)
    
y_train_pred.head()
```

```python
#calculating various metrics for different treshold cutoffs
cutoff_df = pd.DataFrame( columns = ['prob', 'accuracy', 'sensitivity', 'specificity'])

# TP = confusion[1,1] # true positive 
# TN = confusion[0,0] # true negatives
# FP = confusion[0,1] # false positives
# FN = confusion[1,0] # false negatives

for i in numbers:
    cm1 = metrics.confusion_matrix(y_train_pred['Y'], y_train_pred['thresh_' + str(i.round(2))])
    total1=sum(sum(cm1))
    accuracy = (cm1[0,0]+cm1[1,1])/total1
    
    speci = cm1[0,0]/(cm1[0,0]+cm1[0,1])
    sensi = cm1[1,1]/(cm1[1,0]+cm1[1,1])
    cutoff_df.loc[i] =[ i, accuracy, sensi, speci]
cutoff_df
```

```python
# Plotting accuracy sensitivity and specificity for various probabilities.
cutoff_df.plot(kind = 'line', x = 'prob', y = ['accuracy', 'sensitivity', 'specificity'], figsize = [12, 7])

plt.ylim(-0.2, 1.2)
plt.plot([0.38, 0.38], [0.80, 1.0], color = 'red', linestyle = 'dashed')

plt.show()
```

```python
# Function to plot confusion matrix

def plot_ConfusionMatrix_metrics(
    conf_mat, 
    test_y, 
    pred_y, 
    figsize = None, 
    class_names = None, 
    hide_spines = False, 
    hide_ticks = False, 
    title = ''
): 
    '''
    conf_mat: Output of the sklearn's confusion_matrix method
    test_y: The true values of dependent variable
    pred_y: The predicted values of dependent variable
    figsize: The size of confusion matrix plot that needs to be displayed
    class_names: Axes titles
    hide_splines: Option to remove splines. Default False
    hide_ticks: OPtion to hide ticks. Default False
    title: Title of the graph
    '''
    if figsize is None:
        figsize = (len(conf_mat)*3, len(conf_mat)*3)
    
    fig, ax = plt.subplots(figsize = figsize)
    
    cmap = ListedColormap(['c', '#ebab92', '#9b36c9'])
    
    matshow = ax.matshow(conf_mat, cmap = cmap)
    
    for i in range(conf_mat.shape[0]):
        for j in range(conf_mat.shape[1]):
            cell_text = ''
            cell_text += format(conf_mat[i, j], '.0f')
            ax.text(x=j,
                        y=i,
                        s=cell_text,
                        va='center',
                        ha='center',
                        fontsize = 20,
                        color="black")

    if hide_spines:
        ax.spines['right'].set_visible(False)
        ax.spines['top'].set_visible(False)
        ax.spines['left'].set_visible(False)
        ax.spines['bottom'].set_visible(False)
        
    ax.xaxis.set_ticks_position('top')
    
    ax.set_xticklabels(class_names, fontsize = 17)
    ax.set_yticklabels(class_names, fontsize = 17)
    ax.xaxis.set_label_coords(0.5, 1.15)
    
    plt.title(title, fontsize = 20, y = 1.2)
    plt.xlabel('Predicted', fontsize = 15)
    plt.ylabel('Actual', fontsize = 15)
    
    ax.text(2.5, 0.2, 'Accuracy %:' + str(round(accuracy_score(test_y, pred_y) * 100, 2)), fontsize = 17)
    ax.text(1.8, 0.4, 'Precision %: ' + str(round(precision_score(test_y, pred_y) *100, 2)), fontsize = 17)
    ax.text(3.4, 0.4, 'Sensitivity: ' + str(round(conf_mat[1][1] / (conf_mat[1][1] + conf_mat[1][0]), 2)), fontsize = 17)
    ax.text(1.8, 0.7, 'Specificity: ' + str(round(conf_mat[0][0] / (conf_mat[0][0] + conf_mat[0][1]), 2)), fontsize = 17)
    ax.text(3.4, 0.7, 'F1 Score: ' + str(round(f1_score(test_y, pred_y), 2)), fontsize = 17)
    ax.text(2.725, 0.55, 'AUC: ' + str(round(roc_auc_score(test_y, pred_y), 2)), fontsize = 17)
    
    plt.tight_layout()
```

```python
#plotting the confusion matrix for Train dataset
plot_ConfusionMatrix_metrics(confusion_matrix(y_train_pred['Y'], y_train_pred['Result']), 
                             y_train_pred['Y'], y_train_pred['Result'], 
                             class_names = ['0', '1','0','1'], 
                             title = 'Logistic Regression - Train')
```

```python
#function to calculate metrics
def metrics(
    test_y,
    pred_y,
    conf_mat
):
    x = list()
    x.append(round(accuracy_score(test_y, pred_y)*100 , 2))
    x.append(round(precision_score(test_y, pred_y)*100, 2))
    x.append(round(conf_mat[1][1] / (conf_mat[1][1] + conf_mat[1][0]), 2))
    x.append(round(conf_mat[0][0] / (conf_mat[0][0] + conf_mat[0][1]), 2))
    x.append(round(f1_score(test_y, pred_y), 2))
    x.append(round(roc_auc_score(test_y, pred_y), 2))
    return x
```

```python
#calculating train n validation metrics
train_metrics= metrics(y_train_pred['Y'], y_train_pred['Result'], confusion_matrix(y_train_pred['Y'], y_train_pred['Result']))
val_metrics=metrics(y_val_pred['Y'], y_val_pred['Result'], confusion_matrix(y_val_pred['Y'], y_val_pred['Result']))
```
